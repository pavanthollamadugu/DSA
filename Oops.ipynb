```python
 # Inorder to remove the extra code we use init method
 #__init__ is used to initalize values when object is created

 # self parameter --> Object

# from the above the code we can write


class Male:
  def __init__(self,color,weight):
    self.color=color
    self.weight=weight
male1=Male("black", 60)
male2=Male("white",70)
print(male1.color, male1.weight)
print(male2.color, male2.weight)
print(male1.__dict__)
print(male2.__dict__)
```

```python

# Class Methods
# The attributes are the things that the object has and the methods are the behaviour of the objects.
class Youtube:
    def __init__(self, username, subscribers=0, subscriptions=0):
        self.username = username
        self.subscribers = subscribers
        self.subscriptions = subscriptions

    def subscribe(self, user):
        user.subscribers += 1
        self.subscriptions += 1


user1 = Youtube("pavan")
user2 = Youtube("venkata")

user1.subscribe(user2)

print(user1.__dict__)
print(user2.__dict__)
print(Youtube.__dict__)
```
